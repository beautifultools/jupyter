```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('/Users/goodtool/jupyter/data.xlsx')

data.rename(columns={'성별(1)':'성별', '연령별(1)':'연령'}, inplace=True)

data.replace('1세 미만', '0세', inplace=True)
data.replace('1~9세', '1세', inplace=True)
data.replace('10~19세', '10대', inplace=True)
data.replace('20~29세', '20대', inplace=True)
data.replace('30~39세', '30대', inplace=True)
data.replace('40~49세', '40대', inplace=True)
data.replace('50~59세', '50대', inplace=True)
data.replace('60~69세', '60대', inplace=True)
data.replace('70~79세', '70대', inplace=True)
data.replace('80세 이상', '80대', inplace=True)


data.replace('(.*).1', r'\테스트\1', inplace=True, regex=True)
data.replace('.2', '30~2시간', inplace=True)
data.replace('.3', '2~4시간', inplace=True)
data.replace('.4', '6~8시간', inplace=True)
data.replace('.5', '8~12시간', inplace=True)

data_M_30 = data[data['성별'] == '남자'][data['연령'] == '30대']

# 차트 윈도우 제목
plt.figure(num = "30대 남성")

# 차트 글꼴
plt.rcParams.update({'font.family':'malgun gothic', 'font.size':12})

plt.title('30대 남성 데이터')

ax = plt.gca()

print(data_M_30)
data_M_30.plot(kind='line', ax = ax, marker = 'o')

plt.xlabel("2")
plt.ylabel("1")










```
